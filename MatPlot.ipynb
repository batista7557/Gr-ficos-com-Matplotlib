Importação das bibliotecas

```python
import matplotlib.pyplot as plt
import numpy as np
```

Desenhando uma linha diagonal em um gráfico

```python
xpoints = np.array([0,6])
ypoints = np.array([0,250])

plt.plot(xpoints, ypoints)
plt.show()
```

Desenhando um gráfico com posições diferentes

```python
xpoints = np.array([30,6])
ypoints = np.array([10,25])

plt.plot(xpoints, ypoints)
plt.show()
```

Gráfico sem linha

```python
xpoints = np.array([1,8])
ypoints = np.array([3,10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

Gráfico com múltiplos pontos

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints)
plt.show()
```

Marcador

```python
#Se não especificar os pontos do eixo 'x' por exemplo, ele irá seguir dos pontos 1,2,3... em ordem, sucessivamente
ypoints = np.array([3,8,1,10])

plt.plot(ypoints, marker='o')
plt.show()
```

```python
#Substituindo por uma estrela
plt.plot(ypoints, marker='*')
plt.show()
```

Linha pontilhada e adicionando cores

```python
#o - Pontos
#: - Linha pontilhada
#r - cor vermelha
plt.plot(ypoints, 'o:r')
plt.show()
```

```python
#Alterando as dimensões dos pontos - ms
plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

```python
#Cor do contorno do marcador - mec
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

```python
#Cor do marcador - mfc
plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

Estilo da linha

```python
#linestyle - pontilhada
plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
#linestyle - traços
plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

Cor da linha

```python
plt.plot(ypoints, color = 'r')
plt.show()
```

Expessura da linha

```python
plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

Múltiplas linhas

```python
y1 = np.array([3,8,1,10])
y2 = np.array([6,2,7,11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

Labels e título

```python
x = np.array([80, 85, 90])
y = np.array([240, 250, 260])

plt.plot(x, y)

plt.title("Carga X Altura") #Título
plt.xlabel("Carga") #Label do x
plt.ylabel("Altura") #Label do y

plt.show()
```

Alterando as fontes

```python
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Carga X Altura", fontdict = font1)
plt.xlabel("Carga", fontdict = font2)
plt.ylabel("Altura", fontdict = font2)

plt.plot(x, y)
plt.show()
```

Posicionando o título

```python
x = np.array([80, 85, 90])
y = np.array([240, 250, 260])

plt.plot(x, y)

plt.title("Carga X Altura", loc = 'left') #Título à esquerda
plt.xlabel("Carga") #Label do x
plt.ylabel("Altura") #Label do y

plt.show()
```

Adicionando grid

```python
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
#Adicionando propriedades
plt.plot(x, y)
plt.grid(color = 'green', linestyle = '--', linewidth = 1.5)
plt.show()
```

Gerando vários gráficos

```python
#plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #1 linha, 1 coluna e é o gráfico 1
plt.plot(x, y)

#plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) #1 linha, 1 coluna e é o gráfico 2
plt.plot(x, y)

plt.show()
```

```python
#plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1) #2 linha, 1 coluna e é o gráfico 1
plt.plot(x, y)

#plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2) #2 linha, 1 coluna e é o gráfico 2
plt.plot(x, y)

plt.show()
```

```python
#plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #1 linha, 1 coluna e é o gráfico 1
plt.plot(x, y)
plt.title("VENDAS")

#plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2) #1 linha, 1 coluna e é o gráfico 2
plt.plot(x, y)
plt.title("LUCROS")

plt.suptitle("LOJA") #Super title
plt.show()
```

Gráfico de barras

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y) #Na horizontal
plt.show()
```

```python
#Mudando a cor
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "red")
plt.show()
```

```python
#Espessura das colunas
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1)
plt.show()
```
